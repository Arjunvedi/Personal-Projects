```python
import yfinance as yf
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
import seaborn as sns

# Parameters
STOCK_TICKERS = ['TCS.NS', 'INFY.NS']
SEQ_LENGTHS = [5, 10, 20, 25]
EPOCHS = 25
BATCH_SIZE = 32
LEARNING_RATE = 0.001
HIDDEN_SIZE = 50

# Download stock data
def fetch_stock_data(ticker):
    stock = yf.download(ticker, period='5y', interval='1d')
    stock['MA10'] = stock['Close'].rolling(window=10).mean()
    stock['MA50'] = stock['Close'].rolling(window=50).mean()
    stock['Label'] = (stock['Close'].shift(-1) > stock['Close']).astype(np.float32)  # Up = 1.0, Down = 0.0
    stock.dropna(inplace=True)
    return stock

# Prepare dataset
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length, :-1])
        y.append(data[i+seq_length, -1])
    return np.array(X, dtype=np.float32), np.array(y, dtype=np.float32).reshape(-1, 1)

# Model definition
class StockRNN(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(StockRNN, self).__init__()
        self.rnn = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        _, (h_n, _) = self.rnn(x)
        x = self.fc(h_n[-1])
        return self.sigmoid(x).view(-1)

# Training function
def train_model(model, train_loader, criterion, optimizer, epochs):
    model.train()
    losses = []
    accuracies = []
    for epoch in range(epochs):
        epoch_loss = 0
        correct, total = 0, 0
        for X_batch, y_batch in train_loader:
            X_batch = X_batch.to(torch.float32)
            y_batch = y_batch.to(torch.float32).view(-1)
            optimizer.zero_grad()
            outputs = model(X_batch).squeeze()
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()

            # Track accuracy during training
            preds = (outputs > 0.5).int()
            correct += (preds == y_batch.int()).sum().item()
            total += y_batch.size(0)

        epoch_acc = correct / total
        losses.append(epoch_loss / len(train_loader))
        accuracies.append(epoch_acc)
        print(f"Epoch {epoch+1}/{epochs}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.2%}")

    return losses, accuracies

# Evaluation function
def evaluate_model(model, test_loader):
    model.eval()
    predictions, actuals = [], []
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch = X_batch.to(torch.float32)
            outputs = model(X_batch).squeeze()
            preds = (outputs > 0.5).int().numpy()
            predictions.extend(preds)
            actuals.extend(y_batch.numpy().flatten())
    acc = accuracy_score(actuals, predictions)
    prec = precision_score(actuals, predictions)
    rec = recall_score(actuals, predictions)
    f1 = f1_score(actuals, predictions)
    print(f"Test Accuracy: {acc:.2%}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1 Score: {f1:.4f}")
    return actuals, predictions

# Train and test across different sequence lengths for each stock
for ticker in STOCK_TICKERS:
    print(f"\nProcessing Stock: {ticker}")
    stock_data = fetch_stock_data(ticker)
    features = stock_data[['Close', 'MA10', 'MA50']].values
    scaler = StandardScaler()
    features = scaler.fit_transform(features)

    for seq_length in SEQ_LENGTHS:
        print(f"\nTraining with Sequence Length: {seq_length}")
        X, y = create_sequences(np.hstack((features, stock_data[['Label']].values)), seq_length)
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
        train_loader = torch.utils.data.DataLoader(list(zip(X_train, y_train)), batch_size=BATCH_SIZE, shuffle=True)
        test_loader = torch.utils.data.DataLoader(list(zip(X_test, y_test)), batch_size=BATCH_SIZE, shuffle=False)

        model = StockRNN(input_size=3, hidden_size=HIDDEN_SIZE)
        criterion = nn.BCELoss()
        optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

        train_losses, train_accuracies = train_model(model, train_loader, criterion, optimizer, EPOCHS)
        actuals, predictions = evaluate_model(model, test_loader)

        # Plot loss curve
        plt.figure(figsize=(6, 4))
        plt.plot(train_losses, label=f'{ticker} - Seq {seq_length}')
        plt.title(f"Training Loss Curve - {ticker} (Seq {seq_length})")
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
        plt.legend()
        plt.show()

        # Plot accuracy curve
        plt.figure(figsize=(6, 4))
        plt.plot(train_accuracies, label=f'{ticker} - Seq {seq_length}', color='green')
        plt.title(f"Training Accuracy Curve - {ticker} (Seq {seq_length})")
        plt.xlabel("Epochs")
        plt.ylabel("Accuracy")
        plt.legend()
        plt.show()

        # Confusion Matrix
        cm = confusion_matrix(actuals, predictions)
        plt.figure(figsize=(5, 5))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f"Confusion Matrix - {ticker} (Seq {seq_length})")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.show()

        # Display sample predictions
        sample_df = pd.DataFrame({'Actual': actuals[:10], 'Predicted': predictions[:10]})
        print(f"\nSample Predictions for {ticker} (Seq {seq_length}):")
        print(sample_df)
```
